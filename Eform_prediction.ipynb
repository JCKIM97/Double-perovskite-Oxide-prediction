```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
import lightgbm as lgb
from sklearn.metrics import mean_absolute_error
from pandas import DataFrame
from math import sqrt


df1= pd.read_csv("MP_scaled2021.csv", encoding = 'utf-8')
df2= pd.read_csv("DP_scaled2021.csv", encoding = 'utf-8')
df2= df2.dropna()

train_r2 = 0
test_r2 = 0
train_mse = 0
test_mse = 0
train_rmse = 0
test_rmse = 0
train_mae = 0
test_mae = 0

for i in range(0, 100, 10) :
    mp_train_set, mp_val_set = train_test_split(df1.dropna(), test_size=0.2, random_state=i)
    dp_train_set, dp_test_set = train_test_split(df2.dropna(), test_size=0.2, random_state=i)
        
    train_set = mp_train_set.append(dp_train_set)
    test_set = dp_test_set
        
    x_train = train_set.drop(["formation_energy_per_atom", "Index", "formula", "e_above_hull"], axis=1)
    y_train = train_set["formation_energy_per_atom"].copy()
    x_val = mp_val_set.drop(["formation_energy_per_atom", "Index", "formula", "e_above_hull"], axis=1)
    y_val = mp_val_set["formation_energy_per_atom"].copy()
    x_test = test_set.drop(["formation_energy_per_atom", "Index"], axis=1)
    y_test = test_set["formation_energy_per_atom"].copy()
    
    
    gbm = lgb.LGBMRegressor(num_leaves = 1000,
                            num_threads = 16,
                            learning_rate = 0.05,
                            n_estimators = 500,
                            objective = 'regression',)

    gbm.fit(x_train, y_train,
            eval_set=[(x_val, y_val)],
            eval_metric='rmse',                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                              
            early_stopping_rounds=20)



    y_pred = gbm.predict(x_test)
      
            
    R1 = r2_score(list(y_train), gbm.predict(x_train))
    R2 = r2_score(list(y_test), gbm.predict(x_test))
    mse1 = mean_squared_error(list(y_train), gbm.predict(x_train))
    mse2 = mean_squared_error(list(y_test), gbm.predict(x_test))
    mae1 = mean_absolute_error(list(y_train), gbm.predict(x_train))
    mae2 = mean_absolute_error(list(y_test), gbm.predict(x_test))
    rmse1 = sqrt(mse1)
    rmse2 = sqrt(mse2)
        
    train_r2 = train_r2 + R1
    test_r2 = test_r2 +R2
    train_mse = train_mse + mse1
    test_mse = test_mse + mse2
    train_rmse = train_rmse + rmse1
    test_rmse = test_rmse + rmse2
    train_mae = train_mae + mae1
    test_mae = test_mae + mae2

    print('train_r2 :', R1)
    print('test_r2 :', R2)
    print('train_mse :', mse1)
    print('test_mse :', mse2)
    print('train_rmse :', rmse1)
    print('test_rmse :', rmse2)
    print('train_mae :', mae1)
    print('test_mae :', mae2)
    
train_r2 = train_r2 /10
test_r2 = test_r2 /10
train_mse = train_mse/10
test_mse = test_mse/10
train_rmse = train_rmse/10
test_rmse = test_rmse/10
train_mae = train_mae/10
test_mae = test_mae/10

print('average train_r2 : ', train_r2)
print('average test_r2 : ', test_r2)
print('average train_mse : ', train_mse)
print('average test_mse : ', test_mse)
print('average train_rmse : ', train_rmse)
print('average test_rmse : ', test_rmse)
print('average train_mae : ', train_mae)
print('average test_mae : ', test_mae)
```
