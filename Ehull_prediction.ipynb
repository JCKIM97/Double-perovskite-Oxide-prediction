```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
import lightgbm as lgb
from sklearn.metrics import mean_absolute_error
from pandas import DataFrame
from math import sqrt
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
from sklearn.metrics import recall_score
from sklearn.metrics import precision_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix

def score(y_test, pred):
    accuracy = accuracy_score(y_test, pred)
    f1score = f1_score(y_test, pred, average='weighted')
    recall = recall_score(y_test, pred, average='weighted')
    precision = precision_score(y_test, pred, average='weighted')

    print(f'Mean accuracy score: {accuracy:.4}')
    print(f'F1 score: {f1score:.3}')
    print(f'Recall: {recall:.3}')
    print(f'Precision: {precision:.3}')

df1= pd.read_csv("MP_scaled2021.csv", encoding = 'utf-8')
df2= pd.read_csv("DP_scaled2021_ehull.csv", encoding = 'utf-8')

ehull = df1["e_above_hull"]

ehull_classification = []

for i in range(len(ehull)):
    e = ehull[i]
    if e == 0:
        ehull_classification.append('stable')
    elif e < 0.1 :
        ehull_classification.append('meta_stable')
    else :
        ehull_classification.append('instable')
        
df1['ehull_classification'] = ehull_classification


ehull2 = df2["e_above_hull"]

ehull_classification2 = []

for i in range(len(ehull2)):
    e = ehull2[i]
    if e == 0:
        ehull_classification2.append('stable')
    elif e < 0.1 :
        ehull_classification2.append('meta_stable')
    else :
        ehull_classification2.append('instable')
        
df2['ehull_classification'] = ehull_classification2

for i in range(0, 100, 10) :
    mp_train_set, mp_val_set = train_test_split(df1.dropna(), test_size=0.2, random_state=i)
    dp_train_set, dp_test_set = train_test_split(df2.dropna(), test_size=0.9, random_state=i)
 
    train_set = mp_train_set.append(dp_train_set)
    test_set = dp_test_set
        
    x_train = train_set.drop(["formation_energy_per_atom", "Index", "formula", "e_above_hull", "ehull_classification"], axis=1)
    y_train = train_set["ehull_classification"].copy()
    x_val = mp_val_set.drop(["formation_energy_per_atom", "Index", "formula", "e_above_hull", "ehull_classification"], axis=1)
    y_val = mp_val_set["ehull_classification"].copy()
    x_test = test_set.drop(["formation_energy_per_atom", "Index", "e_above_hull", "ehull_classification"], axis=1)
    y_test = test_set["ehull_classification"].copy()
    
    
    gbm = lgb.LGBMClassifier(num_leaves = 1000,
                            num_threads = 16,
                            learning_rate = 0.05,
                             n_estimators = 500,
                            objective = 'regression',)

    gbm.fit(x_train, y_train)


    y_pred = gbm.predict(x_test)
    score(y_test, y_pred)

            
```
